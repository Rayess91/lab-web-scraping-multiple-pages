```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Python'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
links = soup.find_all('a')
```

```python
links_list = [link.get('href') for link in links]

print(links_list)
```
